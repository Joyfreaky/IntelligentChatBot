```python
from azure.identity import DefaultAzureCredential
from azure.keyvault.secrets import SecretClient
from azure.storage.blob import BlobServiceClient
import pandas as pd
import csv
```

```python
# The URL to your Key Vault
key_vault_url = "https://cybersecuritykey.vault.azure.net/"

# The name of your secret
secret_name = "cybersecurityDataConnectionString"

# The name of your blob container
container_name = "cybersecuritydata"

# The name of the blob
blob_name = "dataset_subset.csv"

# Create a credential object using the DefaultAzureCredential class
credential = DefaultAzureCredential()

# Create a secret client using the credential
secret_client = SecretClient(vault_url=key_vault_url, credential=credential)

# Retrieve the secret
secret = secret_client.get_secret(secret_name)

# Create a blob service client using the secret
blob_service_client = BlobServiceClient.from_connection_string(secret.value)

# Get a reference to the blob container
blob_container_client = blob_service_client.get_container_client(container_name)

# Get a reference to the blob
blob_client = blob_container_client.get_blob_client(blob_name)
```

```python
# Download the blob to a local file
with open("dataset_subset.csv", "wb") as my_blob:
    download_stream = blob_client.download_blob()
    my_blob.write(download_stream.readall())

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv("dataset_subset.csv")

# Print the DataFrame
df.head(10)
```

```python
# Drop the first unnamed column if it's an index column
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)

df.head(10)
```

```python
# Handle missing values by replacing them with a placeholder string
df.fillna('Unknown', inplace=True)

df.head(10)
```

```python
# Convert lists represented as strings to actual lists (if necessary)
# For example, if SourceIP is a string that looks like "['1.2.3.4']", we need to evaluate it as a list
import ast
list_columns = ['SourceIP', 'TargetIP', 'NodeType', 'SourcePort', 'TargetPort', 'SourceProtocol', 'TargetProtocol']
for col in list_columns:
    df[col] = df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x)

df.head(10)
```

```python
label_counts = df['Category'].value_counts()
print(label_counts)
```

```python
# We will drop those rows who's labels count is less than 10

labels_to_drop = label_counts[label_counts < 10].index

df = df[~df['Category'].isin(labels_to_drop)]

print(len(df))
```

```python
"""# We will also reduce the number of rows for randomly, for Recon.Scanning, Attempt.Login, Attempt.Exploit to remove Bais from the training set.
# Define the categories and the desired percentages
categories_to_reduce = {
    'Recon.Scanning': 0.99,
    'Attempt.Login': 0.9,
    'Attempt.Exploit': 0.9,
}

# Create a mask to identify rows for the specified categories
mask = df['Category'].isin(categories_to_reduce.keys())

# Apply the reduction based on the defined percentages
for category, percentage in categories_to_reduce.items():
    category_rows = df[df['Category'] == category]
    num_rows_to_keep = int(len(category_rows) * (1 - percentage))
    category_rows_to_keep = category_rows.sample(n=num_rows_to_keep, random_state=42)
    df = pd.concat([df[df['Category'] != category], category_rows_to_keep])"""

# Print the updated category counts
print(df['Category'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Convert categorical labels to numerical labels
label_encoder = LabelEncoder()
df['Category_encoded'] = label_encoder.fit_transform(df['Category'])

import joblib
# Save the label encoder as a dictionary

label_encoder_dict = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
joblib.dump(label_encoder_dict, 'label_encoder_dict.joblib')

df.head(10)
```

```python
# Concatenate the columns into a single feature column for BERT

df['feature'] = (
    "Detect Time: " + df['DetectTime'].astype(str) + "; " +
    "Event time: " + df['EventTime'].astype(str) + "; " +
    "Source IP: " + df['SourceIP'].astype(str) + "; " +
    "Target IP: " + df['TargetIP'].astype(str) + "; " +
    "Source Port: " + df['SourcePort'].astype(str) + "; " +
    "Target Port: " + df['TargetPort'].astype(str) + "; " +
    "Source Protocol: " + df['SourceProtocol'].astype(str) + "; " +
    "Target Protocol: " + df['TargetProtocol'].astype(str) + "; " +
    "Win Start Time: " + df['WinStartTime'].astype(str) + "; " +
    "Win End Time: " + df['WinEndTime'].astype(str) + "; " +
    "Connection Count: " + df['ConnCount'].astype(str) + "; " +
    "Flow Count: " + df['FlowCount'].astype(str) + "; " +
    "Node Name: " + df['NodeName'].astype(str) + "; " +
    "Node Type: " + df['NodeType'].astype(str) + "; "
)

df['feature'].head(10)
```

```python
# Create a new column 'Annotation' and add annotations to each row
annotations_str = "; ".join([f"{k}: {v}" for k, v in feature_annotations.items()])
df['Annotation'] = annotations_str

# Concatenate the existing 'feature' column with the new 'Annotation' column
df['feature'] = df['feature'] + "; Annotation: " + df['Annotation']

df['feature'].head(10)
```

```python
# List of columns to keep
columns_to_keep = ['Category_encoded', 'feature', 'ThreatAnnotation']

# Create a new DataFrame with only the selected columns
df = df[columns_to_keep]

# Rename the Category_encoded to labels
df.rename(columns={'Category_encoded': 'labels'}, inplace=True)

# Display the first few rows of the final DataFrame
df.head(10)

```

```python
# Prepare the dataset for training and validation
from sklearn.model_selection import train_test_split

# Specify the columns for the input text and the target labels
input_text_column = 'feature'
target_label_column = 'labels'


# Split the data into train, validation, and test sets
train_df, test_valid_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df[target_label_column])
valid_df, test_df = train_test_split(test_valid_df, test_size=0.5,random_state=42, stratify=test_valid_df[target_label_column])

# Drop 'ThreatAnnotation' from validation and test sets
valid_df = valid_df.drop(['ThreatAnnotation'], axis=1)
test_df = test_df.drop(['ThreatAnnotation'], axis=1)

# Concatenate 'feature' with 'ThreatAnnotation' for the training set
train_df['feature'] = train_df['feature'] + "; ThreatAnnotation: " + train_df['ThreatAnnotation']

# Drop 'ThreatAnnotation' for the training set
train_df = train_df.drop(['ThreatAnnotation'], axis=1)

# Display the first few rows of each set
print("Train Set:")
print(train_df.head())
print("\nValidation Set:")
print(valid_df.head())
print("\nTest Set:")
print(test_df.head())
```
